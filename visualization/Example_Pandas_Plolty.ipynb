## Example: Regioni totale case attuali e incremento di oggi (04-20-2022)


### Create table and save data as excel-csv

```python
# use pandas for the table filter by datetime
import pandas as pd
from datetime import datetime

dati_regioni = "https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv"
df = pd.read_csv(dati_regioni)
#display( df.tail(1) )
dfn = df[["data","denominazione_regione","totale_casi"]]
dfn["data"] = pd.to_datetime(dfn['data'], format='%Y-%m-%dT%H:%M:%S')
# dfn = dfn[ ( (dfn["data"]>datetime(2022,4,18,12,0,0)) & (dfn["data"]<datetime(2022,4,20)) ) ]

# collect data of yesterday and data of today
yesterday = dfn[ ( ( dfn["data"]>datetime(2022,4,19,0,0,0) ) & (  dfn["data"]<datetime(2022,4,20,0,0,0) ) )   ]["totale_casi"]
today = dfn[ ( ( dfn["data"]>datetime(2022,4,20,0,0,0) ) & (  dfn["data"]<datetime(2022,4,21,0,0,0) ) )   ]["totale_casi"]

# select only data of today from dataframe
dfn = dfn[ dfn["data"]>datetime(2022,4,20,12,0,0) ]

# add columns increment
dfn.insert(3,"increment",today.values-yesterday.values)

# display table
display(dfn)

# save table
dfn.to_excel("result.xlsx", index=None)
dfn.to_csv("result.csv", index=None)
dfn.to_html("result.html", index=None)
#dfn.to_latex(index=None)
```

### Convert pandas table to plotly figure

* [Figure factory plolty](https://plotly.com/python/figure-factory-table/)
* [Figure Referece plotly](https://plotly.com/python/reference/)

```python
import plotly.figure_factory as ff
fig = ff.create_table(dfn)
fig.show()
fig.write_image("table.svg")
fig.write_image("table.png")
fig.write_image("table.pdf")
```

## Plotting example

* [Emulate di website](https://opendatadpc.maps.arcgis.com/apps/dashboards/dae18c330e8e4093bb090ab0aa2b4892)

```python
import plotly.express as px
import pandas as pd
dati_italia = "https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv"
df = pd.read_csv(dati_italia)
fig = px.line(df,x="data", y = ["dimessi_guariti","totale_positivi"],title="Andamento nazionale")
fig.update_layout(
    title_font_color="white",
    paper_bgcolor="black",
    plot_bgcolor="black",
    hovermode="x"
)
fig.show()
```

```python
fig = px.line(df,x="data", y = ["nuovi_positivi"],title="Nuovi Positivi")
fig.show()
```

```python
fig = px.area(df,x="data", y = ["variazione_totale_positivi"],title="Variazione Totale Positivi")
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_traces( go.Scatter(x=df["data"], y = df["variazione_totale_positivi"], mode="markers+lines"))
fig.update_layout(
    title="Variazione totale positivi",
    title_font_color="white",
    paper_bgcolor="black",
    plot_bgcolor="black",
    hovermode="x"
)
fig.show()
```

```python
fig.write_html("fig.html")
fig.write_image("fig.png")
fig.write_image("fig.svg")
```
