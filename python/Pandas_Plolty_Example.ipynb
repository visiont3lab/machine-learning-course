```python
import plotly.graph_objects as go
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv")
df.to_csv("myfile.csv")
```

```python
def get_data(nome_della_colonna):
  return df[nome_della_colonna].values[-1]

def get_incremento(nome_della_colonna):
  return df[nome_della_colonna].values[-1] - df[nome_della_colonna].values[-2]
```

```python
print(f'Totale Positivi: %s ' % ( get_data('totale_positivi') ) )
print(f'Totale Positivi Inc: %s ' % ( get_incremento('totale_positivi') ) )
print(f'Deceduti: %s ' % ( get_data('deceduti') ) )
print(f'Deceduti Inc: %s ' % ( get_incremento('deceduti') ) )
```

## Plot

```python
def plot_fig(df,nome_colonna):
  fig = go.Figure()
  x = df["data"] #[4,5,6]
  fig.add_trace( go.Scatter( x=x, y=df[nome_colonna]) )
  fig.update_layout(title=nome_colonna)
  fig.show()

plot_fig(df,"deceduti")
plot_fig(df,"totale_positivi")
```
