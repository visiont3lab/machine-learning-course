```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
X = iris.data 
```

```python
print(X.shape)
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
pca.components_
```

```python
Z = pca.transform(X)
```

```python
Z.shape
```

```python
 plt.scatter(Z[:,0],Z[:,1]); # questo Ã¨ l'iperpiano in 2 dimensioni che massimizza la varianza
```

```python
#lo coloriamo per classi, ponendo come colore=y
y = iris.target 
```

```python
plt.figure(figsize=(18,10))
plt.scatter(Z[:,0],Z[:,1],c=y);
plt.xlabel('PC1')
plt.ylabel('PC2')
```
