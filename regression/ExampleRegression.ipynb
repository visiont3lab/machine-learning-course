```python
from sklearn import datasets
import pandas  as pd

# Dataset diabetes
data = datasets.load_diabetes()
X = data["data"]
Y = data["target"].reshape(-1,1)

# Dataset boston
#data_url = "http://lib.stat.cmu.edu/datasets/boston"
#raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
#X = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
#Y = raw_df.values[1::2, 2]
```

```python
from sklearn import linear_model
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
import numpy as np

# Standard Scalare (normalizza i dati X-mean/std)
c = StandardScaler()
X = c.fit_transform(X)

# Polynomial featuers
#c = PolynomialFeatures(degree=2)
#X = c.fit_transform(X)

# Train test split (divide i dati in train e test)
X_train, X_test,  Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=1)

# Scelta modello
m = linear_model.LinearRegression()
#m = ExtraTreesRegressor()
#m = GradientBoostingRegressor()
#m = RandomForestRegressor()

# Trrain
m.fit(X_train,Y_train)

# Error train
Y_hat_train = m.predict(X_train)
err_train = np.sqrt( mean_squared_error(Y_hat_train,Y_train) )
print("Risultato train: ", err_train)

# Error test
Y_hat_test = m.predict(X_test)
err_test = np.sqrt( mean_squared_error(Y_hat_test,Y_test) )
print("Risultato test: ", err_test)

# Plot result train
f = go.Figure()
f.add_trace(go.Scatter(x=np.linspace(0,1,Y_train.shape[0]),y=Y_train.flatten(), mode="lines+markers", name="original"))
f.add_trace(go.Scatter(x=np.linspace(0,1,Y_hat_train.shape[0]),y=Y_hat_train.flatten(),mode="lines+markers",name="predicted"))
f.show()

# Plot result test
f = go.Figure()
f.add_trace(go.Scatter(x=np.linspace(0,1,Y_test.shape[0]),y=Y_test.flatten(), mode="lines+markers", name="original"))
f.add_trace(go.Scatter(x=np.linspace(0,1,Y_hat_test.shape[0]),y=Y_hat_test.flatten(),mode="lines+markers",name="predicted"))
f.show()
```
