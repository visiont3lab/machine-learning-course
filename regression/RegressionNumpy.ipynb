# Linear Regression Numpy

## Linear Regression Manually set params

```python
import numpy as np
import plotly.graph_objects as go 

'''
Y = Salario al mese in euro
X1 = EtÃ  del lavoratore
Equazione = Y = b + a*X --> Y = w0 + w1*X1
1) Marco:    Y=1100   X1=19  Y=w0*1+w1*X1 --> 1100 = w0*1 + w1*19
2) Daniele:  Y=1150   X1=21  Y=w0*1+w1*X1 --> 1150 = w0*1 + w1*21
3) Davide:   Y=1155   X1=22  Y=w0*1+w1*X1 --> 1155 = w0*1 + w1*22 
4) Marta:    Y=1170   X1=23  Y=w0*1+w1*X1 --> 1170 = w0*1 + w1*23
6) Alessia:  Y=1200   X1=26  Y=w0*1+w1*X1 --> 1200 = w0*1 + w1*26
9) Stella:   Y=1750   X1=33  Y=w0*1+w1*X1 --> 1750 = w0*1 + w1*33
10) Chiara   Y=1640   X1=29  Y=w0*1+w1*X1 --> 1640 = w0*1 + w1*29 
'''

X1 = np.array([ 19,21,22,23,26,33,29 ])
Y = np.array([1100,1150,1155,1170,1200,1750,1640])


w0 = 56.21
w1 = 50.70

Y_hat = w0 + w1*X1 # Regressione lineare  lasso ,elastinnet , 

fig = go.Figure()
fig.add_traces( go.Scatter( x=X1,y=Y, mode="markers", name="Real" ))
fig.add_traces( go.Scatter( x=X1,y=Y_hat, mode="markers", name="Predicted" ))
fig.update_layout(
    hovermode="x"
)
fig.show()
```

## Linear Regression closed form solution


```python
import numpy as np

# numero sample x numero input
X = np.array(
            [
              [1, 19],
              [1, 21],
              [1, 22],
              [1, 23],
              [1, 26],
              [1, 33],
              [1, 29],
            ]
    )

# numero di sample per numero di output
Y = np.array( [
                [1100],
                [1150],
                [1155],
                [1170],
                [1200],
                [1750],
                [1640],
              ]
            )

#W = (XT*X)-1XTY
W = np.matmul( np.linalg.pinv( np.matmul( X.T, X ) ) , np.matmul( X.T, Y ) )
W
```

## Esempio Linear Regression Numpy + Polynomial Feature trick

---



```python
import numpy as np

a = 0.1*np.random.random(1000)
b = -0.1*np.random.random(1000)


X1  = np.linspace(0,20,1000)
Y = np.cos(X1)*np.exp(-0.2*X1) +a + b
Y = Y.reshape(-1,1)

X = np.array([ np.ones(len(X1)), X1, X1**2, X1**3, X1**4,X1**5, X1**6,X1**7, X1**8,X1**9,X1**10]).T #,np.power(X1,3),np.power(X1,4),np.power(X1,5),np.power(X1,6),np.power(X1,7),np.power(X1,8),np.power(X1,9)]).T
W = np.matmul( np.linalg.inv( np.matmul( X.T, X ) ) , np.matmul( X.T, Y ) )
Y_hat = np.matmul(W.T, X.T).T


fig = go.Figure()
fig.add_traces( go.Scatter( x=X1,y=Y[:,0], mode="markers", name="reak" ))
fig.add_traces( go.Scatter( x=X1,y=Y_hat[:,0], mode="markers", name="prediction" ))
fig.update_layout(
    hovermode="x"
)
fig.show()
```

## Esercizio Linear Regression Iterative

```python
import numpy as np

a = 0.1*np.random.random(1000)
b = -0.1*np.random.random(1000)

fig = go.Figure()

X1  = np.linspace(0,20,1000)
Y = np.cos(X1)*np.exp(-0.2*X1) +a + b
Y = Y.reshape(-1,1)

Y_hat_f = []
l = 30
for i in range(0,len(X1),l):
  Xt = X1[i:i+l]
  Yt = Y[i:i+l]
  X = np.array([ np.ones(len(Xt)), Xt]).T # np.power(X1,2),np.power(X1,3),np.power(X1,4),np.power(X1,5),np.power(X1,6),np.power(X1,7),np.power(X1,8),np.power(X1,9)]).T
  #print(X.shape,Yt.shape)
  W = np.matmul( np.linalg.inv( np.matmul( X.T, X ) ) , np.matmul( X.T, Yt ) )
  Y_hat = np.matmul(W.T, X.T).T
  Y_hat_f.extend(Y_hat.reshape(-1))
  fig.add_traces( go.Scatter( x=Xt,y=Y_hat.reshape(-1), mode="markers" ))

#print(Y_hat_f)

fig.add_traces( go.Scatter( x=X1,y=Y[:,0], mode="markers", name="res" ))
fig.update_layout(
    hovermode="x"
)
fig.show()
```
